```python
import discord
import pandas as pd
import nest_asyncio
import os
from dotenv import load_dotenv

nest_asyncio.apply()

load_dotenv()

TOKEN = os.getenv('DISCORD_TOKEN')
CHANNEL_ID = 1278882223499645020 # Reemplaza con el ID numérico de tu canal
NOMBRE_ARCHIVO = 'reporte_chats.csv'
```

```python
# 1. Configurar los Intents (necesario para leer el texto de los mensajes)
intents = discord.Intents.default()
intents.message_content = True

# 2. Inicializar el cliente
client = discord.Client(intents=intents)

# Configurar intents (necesario message_content para leer el historial)
intents = discord.Intents.default()
intents.message_content = True
client = discord.Client(intents=intents)

@client.event
async def on_ready():
    print(f'Conectado como {client.user}. Iniciando extracción...')
    canal = client.get_channel(CHANNEL_ID)
    
    if canal is not None:
        datos_mensajes = []
        
        async for mensaje in canal.history(limit=500):
            # Inicializamos los valores por defecto (asumiendo que NO es respuesta)
            es_respuesta = 0
            id_mensaje_referencia = "N/A"
            usuario_respondido = "N/A"
            
            # Verificamos si el mensaje es una respuesta válida
            if mensaje.reference is not None and mensaje.reference.message_id is not None:
                es_respuesta = 1
                id_mensaje_referencia = mensaje.reference.message_id
                
                # Intentamos obtener los datos del mensaje al que se responde
                msg_referenciado = mensaje.reference.resolved
                
                if isinstance(msg_referenciado, discord.Message):
                    # Si el mensaje existe, guardamos el nombre del autor original
                    usuario_respondido = msg_referenciado.author.name
                elif isinstance(msg_referenciado, discord.DeletedReferencedMessage):
                    # Si el mensaje al que se respondió fue eliminado
                    usuario_respondido = "Mensaje Eliminado"

            # Guardamos la información en la lista
            datos_mensajes.append([
                mensaje.id,
                mensaje.created_at.strftime("%Y-%m-%d %H:%M:%S"),
                mensaje.author.name,
                mensaje.content,
                es_respuesta,
                id_mensaje_referencia,
                usuario_respondido     # Nueva columna con el nombre del usuario
            ])   
        
        # Estructuramos el DataFrame con las nuevas columnas
        columnas = [
            'ID_Mensaje', 'Fecha', 'Autor', 'Contenido', 
            'Es_Respuesta', 'ID_Mensaje_Respondido', 'Usuario_Respondido'
        ]
        df = pd.DataFrame(datos_mensajes, columns=columnas)
        
        # Guardar a CSV
        df.to_csv(NOMBRE_ARCHIVO, index=False, encoding='utf-8')
        print(f"✅ Extracción finalizada. Archivo '{NOMBRE_ARCHIVO}' guardado con éxito ({len(df)} mensajes).")
    else:
        print("❌ Error: No se pudo encontrar el canal.")
        
    # Cerrar la conexión de forma segura
    await client.close()

# Ejecutar el bot para la Celda 1
client.run(TOKEN)
```

```python
# 1. Volvemos a instanciar el cliente usando los 'intents' de la memoria
client = discord.Client(intents=intents)

# 2. Definimos el evento para esta acción específica
@client.event
async def on_ready():
    print(f'Conectado nuevamente. Preparando envío de reporte...')
    
    # CHANNEL_ID ya está en la memoria de Jupyter
    canal = client.get_channel(CHANNEL_ID) 
    
    if canal is not None:
        # NOMBRE_ARCHIVO también está en memoria
        if os.path.exists(NOMBRE_ARCHIVO):
            archivo_discord = discord.File(NOMBRE_ARCHIVO)
            mensaje_texto = "HOLA, acá está lo que han dicho hasta ahora junto a quién le respondieron!!!"
            
            await canal.send(content=mensaje_texto, file=archivo_discord)
            print("✅ Mensaje y archivo adjunto enviados correctamente.")
        else:
            print(f"❌ Error: No se encontró '{NOMBRE_ARCHIVO}'.")
    else:
        print("❌ Error: No se pudo encontrar el canal.")
        
    # Volvemos a cerrar este nuevo cliente al terminar
    await client.close()

# 3. Ejecutamos usando el TOKEN de la memoria
client.run(TOKEN)
```
